```python
import pandas as pd

# Data biodata mahasiswa
df_biodata = pd.DataFrame({
    'npm': [210001, 210002, 210003],
    'nama': ['Adompo', 'Ayyash', 'Dian'],
    'jurusan': ['Informatika', 'Sistem Informasi', 'Informatika']
})

df_biodata
```

```python
# Data nilai mahasiswa
df_nilai = pd.DataFrame({
    'npm': [210001, 210001, 210002, 210003],
    'mata_kuliah': ['Data Mining', 'PBO', 'PBO', 'IMK'],
    'nilai': [90, 85, 80, 88]
})

df_nilai
```

```python
# Integrasi data berdasarkan kolom 'npm'
df_gabungan = pd.merge(df_nilai, df_biodata, on='npm', how='left')
df_gabungan
```
