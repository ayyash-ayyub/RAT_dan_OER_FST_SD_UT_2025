```python
import pandas as pd

# Contoh data awal
data = {
    'fitur1': [10, 20, 30, 40, 50],
    'fitur2': [100, 150, 200, 250, 300]
}

df = pd.DataFrame(data)
print("Data sebelum scaling:")
print(df)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df[['fitur1', 'fitur2']])
df_scaled = pd.DataFrame(df_scaled, columns=['fitur1', 'fitur2'])

print("Data setelah Min-Max Scaling:")
print(df_scaled)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Data sebelum scaling
data = {
    'fitur1': [10, 20, 30, 40, 50],
    'fitur2': [100, 150, 200, 250, 300]
}
df = pd.DataFrame(data)

# Scaling data dengan Min-Max Scaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)
df_scaled = pd.DataFrame(scaled_data, columns=['fitur1', 'fitur2'])

# Plot grafik sebelum dan sesudah scaling
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Grafik A: Data sebelum scaling
axs[0].plot(df['fitur1'], label='fitur1', marker='o')
axs[0].plot(df['fitur2'], label='fitur2', marker='s')
axs[0].set_title('A. Data Sebelum Scaling')
axs[0].set_xlabel('Index')
axs[0].set_ylabel('Nilai Asli')
axs[0].legend()
axs[0].grid(True)

# Grafik B: Data setelah Min-Max Scaling
axs[1].plot(df_scaled['fitur1'], label='fitur1 (scaled)', marker='o')
axs[1].plot(df_scaled['fitur2'], label='fitur2 (scaled)', marker='s')
axs[1].set_title('B. Data Setelah Min-Max Scaling')
axs[1].set_xlabel('Index')
axs[1].set_ylabel('Nilai Skala [0-1]')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# A. Data sebelum standardization
data = {
    'fitur1': [10, 20, 30, 40, 50],
    'fitur2': [100, 150, 200, 250, 300]
}
df = pd.DataFrame(data)

# B. Data setelah standardization
scaler = StandardScaler()
df_standardized = scaler.fit_transform(df)
df_standardized = pd.DataFrame(df_standardized, columns=['fitur1', 'fitur2'])

# Plot perbandingan sebelum dan sesudah standardization
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Grafik A: Data sebelum Standardization
axs[0].plot(df['fitur1'], label='fitur1', marker='o')
axs[0].plot(df['fitur2'], label='fitur2', marker='s')
axs[0].set_title('A. Data Sebelum Standardization')
axs[0].set_xlabel('Index')
axs[0].set_ylabel('Nilai Asli')
axs[0].legend()
axs[0].grid(True)

# Grafik B: Data setelah Standardization
axs[1].plot(df_standardized['fitur1'], label='fitur1 (standardized)', marker='o')
axs[1].plot(df_standardized['fitur2'], label='fitur2 (standardized)', marker='s')
axs[1].set_title('B. Data Setelah Standardization')
axs[1].set_xlabel('Index')
axs[1].set_ylabel('Nilai Z-score')
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```
